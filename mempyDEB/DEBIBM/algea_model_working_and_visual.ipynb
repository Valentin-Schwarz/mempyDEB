```python
sim = simulate_DEBIBM(p) # runs the baseline model (a variant of DEBkiss) with default parameters
mout, aout = sim
# Visualisierung mit Seaborn
plt.figure(figsize=(18, 6))

# Plot für X
plt.subplot(1, 3, 1)  # 1 Zeile, 3 Spalten, 1. Subplot
sns.lineplot(data=mout, x='t_day', y='X', marker='o', color='b')
plt.title('Algenpopulation X über die Zeit')
plt.xlabel('Tage')
plt.ylabel('X')
plt.grid()

# Plot für Q
plt.subplot(1, 3, 2)  # 1 Zeile, 3 Spalten, 2. Subplot
sns.lineplot(data=mout, x='t_day', y='Q', marker='o', color='g')
plt.title('Nährstoffquota Q über die Zeit')
plt.xlabel('Tage')
plt.ylabel('Q')
plt.grid()

# Plot für P
plt.subplot(1, 3, 3)  # 1 Zeile, 3 Spalten, 3. Subplot
sns.lineplot(data=mout, x='t_day', y='P', marker='o', color='r')
plt.title('Phosphor P über die Zeit')
plt.xlabel('Tage')
plt.ylabel('P')
plt.grid()

# Layout anpassen und angezeigt
plt.tight_layout()
plt.show()
```
